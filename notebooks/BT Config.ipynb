# ConfigPreset Backtesting

```python
"""
Phase where you can trial and error, plug and play model with any strategy, until get the best results.
In this phase, feel free to combine the trained model with a variety of technical indicators or ensemble with another model.

TIP: High threshold no trendline |OR| Low threshold with trendline
"""

def strategy_code():
    """ Start Strategy Code """
    from freqtrade.nbtools.strategy import INbStrategy
    from freqtrade.nbtools.remote_utils import load_lightning_container
    from freqtrade.ml.container import LightningContainer
    from pandas import DataFrame
    from functools import reduce
    from copy import deepcopy
    import freqtrade.vendor.qtpylib.indicators as qtpylib
    import numpy as np  # noqa
    import pandas as pd  # noqa
    import talib.abstract as ta
    import pandas_ta as pta
    
    NEXT_CANDLES = 2
    CONTAINERS = [
        ("15n15-catboost_l1", "v0"),
        ("15n60-catboost_l1", "v0"),
        ("15n30-catboost_l3", "v6"),
        ("15n30-catboost_l3", "v5"),
        # ("15n30-catboost_l3", "v4"),
    ]
    MODELS: dict[str, LightningContainer] = {}
    
    for name, ver in CONTAINERS:
        MODELS[name + ":" + ver] = load_lightning_container(name, name, version=ver)

    class NotebookStrategy(INbStrategy):
        # These attributes are REQUIRED!
        timeframe = "15m"
        minimal_roi = {str(NEXT_CANDLES * 15): -1}
        stoploss = -0.99
        startup_candle_count: int = 400
        
        def populate_indicators(self, dataframe: DataFrame, metadata: dict) -> DataFrame:
            for name, model in MODELS.items():
                cols_old = dataframe.columns
                dataframe_preds = model.predict(dataframe)
                dataframe_preds = dataframe_preds.loc[:, [col for col in dataframe_preds.columns if col not in cols_old]]
                dataframe_preds.columns = [f"{name}/{col}" for col in dataframe_preds.columns]
                dataframe[dataframe_preds.columns] = dataframe_preds
            
            # print(dataframe.columns)
            num_classes = 5
            # dataframe["argmax_ml"] = dataframe[[f"ml_{x}"for x in range(num_classes)]].idxmax(axis=1)
            
            dataframe["ema"] = pta.ema(dataframe, length=200)
            dataframe["ema_volume"] = pta.ema(dataframe["volume"], length=50)
            dataframe["rsi_2"] = pta.rsi(dataframe["close"], length=2)
            dataframe["rsi_3"] = pta.ema(dataframe["close"], length=3)
            dataframe["rsi_4"] = pta.ema(dataframe["close"], length=4)
            
            return dataframe

        def populate_buy_trend(self, dataframe: DataFrame, metadata: dict) -> DataFrame:
            # TODO: 
            # - Ensemble minimum votes by count votes for minimum threshold
            # - Adjust minimum threshold to be considered as vote
            
            conditions = [
                # NOTE: Start ML signal #
                dataframe['15n60-catboost_l1:v0/ml_4'] > 0.5,
                dataframe['15n30-catboost_l3:v6/ml_4'] > 0.5,
                dataframe['15n30-catboost_l3:v5/ml_4'] > 0.5,
                dataframe['15n15-catboost_l1:v0/ml_4'] > 0.5,
                # dataframe['15n30-catboost_l3:v4/ml_4'] > 0.55,
                
                # NOTE: Start TA filter #
                # dataframe["close"] > dataframe["ema"],
                dataframe['volume'] > 0,
            ]
            
            dataframe.loc[reduce(lambda x, y: x & y, conditions), 'buy'] = 1
            return dataframe

        def populate_sell_trend(self, dataframe: DataFrame, metadata: dict) -> DataFrame:
            conditions = [
                dataframe['buy'].shift(NEXT_CANDLES) == 1,
                dataframe['buy'] != 1,
                dataframe['volume'] > 0,
            ]
            dataframe.loc[reduce(lambda x, y: x & y, conditions), 'sell'] = 1
            
            return dataframe
    """ End Strategy Code """


preset = ConfigPreset(
    name="catboost-ensemble",
    config_dict = configs.DEFAULT,
    path_data=PATH_LOCAL_DATADIR,
    timerange="20210701-"
)
preset.overwrite_config(pairs=PAIRS_HIGHCAP_NONSTABLE)

stats, summary = backtest(preset, strategy_code, dataloader=DATALOADER)
trades = pd.DataFrame(stats["strategy"]["NotebookStrategy"]["trades"])
plotting.plot_profits_timerange(trades, timerange=preset.timerange, path_mount=PATH_MOUNT, name=preset.name)
```
